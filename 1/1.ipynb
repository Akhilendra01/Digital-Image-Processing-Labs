```python
import numpy as np 
from PIL import Image
import matplotlib.pyplot as plt
import warnings
```

```python
image_path="./mount_fuji.jpeg"
warnings.filterwarnings("ignore")
```

```python
img=Image.open(image_path)
```

```python
img.show()
```

### Resize

```python
new_size=(500, 300)
resized_img=img.resize(new_size)
resized_img.save("resize.jpeg")
resized_img.show()
```

### Grayscale

```python
grayscale_img=img.convert('L')
grayscale_img.save("./graysclae.jpeg")
grayscale_img.show()
```

### Black and white

```python
threshold=128
black_white_img=grayscale_img.point(lambda pixel: 0 if pixel < threshold else 255, '1')
black_white_img.save("./black_and_white.jpeg")
black_white_img.show()
```

### Image profile

```python
width, height = grayscale_img.size
horizontal_profile = [sum(grayscale_img.getpixel((x, y)) for y in range(height)) / height for x in range(width)]

# Plot the profile curve
plt.figure(figsize=(8, 4))
plt.plot(horizontal_profile, color='black')
plt.title('Horizontal Profile Curve')
plt.xlabel('Pixel Position')
plt.ylabel('Pixel Value')
plt.grid(True)
plt.show()
```

### Seperate RGB planes

```python
red, green, blue = img.split()
empty=Image.new('L', img.size, "black")
red=Image.merge("RGB", (red, empty, empty))
green=Image.merge("RGB", (empty, green, empty))
blue=Image.merge("RGB", (empty, empty, blue))
red.save("./red.jpeg")
green.save("./green.jpeg")
blue.save("./blue.jpeg")
red.show()
blue.show()
green.show()
```
